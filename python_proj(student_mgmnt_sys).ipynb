# 📚 Student Management System

A console-based Python project to manage student records.

## Features
- ✅ Add Student
- ✅ View All Students
- ✅ Search Student
- ✅ Update Marks
- ✅ Delete Student
- ✅ Save Data to File
- ✅ Load Data from File

## How it works 
- All student data is stored in a **list of dictionaries**
- Each student has 3 fields -> 'rol_no','name','marks'
- Data can be saved to a '.txt' file and loaded back

## Tools used
- Python built-in functions 
- File handling (open,read,write)
- Exception handling (try/except)

```python
# this list holds all students records
students = []

print("student list initialized")
print(f"current students: {students}") 
```

```python
# Add Student Function

def add_student():
    """Add a new student to the system"""
    print("\n" + "="*50)
    print("ADD STUDENT")
    print("="*50)

    roll_no = input("Enter Roll Number: ").strip()

    # Check if roll number already exists
    for student in students:
        if student["roll_no"] == roll_no:
            print("Roll number already exists!")
            return

    name = input("Enter Name: ").strip()

    try:
        marks = float(input("Enter Marks: "))
        if marks < 0 or marks > 100:
            print("Marks should be between 0-100!")
            return
    except ValueError:
        print("Please enter valid marks!")
        return

    student = {
        "roll_no": roll_no,
        "name":    name,
        "marks":   marks
    }

    students.append(student)
    print("Student Added Successfully!")

print("add_student() defined")
```

```python
# CELL 5 — View Students Function

def view_students():
    # Display all students
    if len(students) == 0:
        print("No students found!")
        return

    print("\n" + "="*50)
    print("ALL STUDENTS")
    print("="*50)
    print(f"{'Roll No':<10} {'Name':<20} {'Marks':<10}")
    print("-"*50)

    for student in students:
        print(f"{student['roll_no']:<10} "
              f"{student['name']:<20} "
              f"{student['marks']:<10}")

    print("="*50)
    print(f"Total Students: {len(students)}")

print("view_students() defined")
```

```python
# Search Student Function

def search_student():
    # Search a student by roll number
    roll_no = input("Enter Roll Number to search: ").strip()

    for student in students:
        if student["roll_no"] == roll_no:
            print("\n" + "="*50)
            print("STUDENT FOUND!")
            print("="*50)
            print(f"Roll No : {student['roll_no']}")
            print(f"Name    : {student['name']}")
            print(f"Marks   : {student['marks']}")
            print("="*50)
            return

    print("Student not found!")

print("search_student() defined")
```

```python
# Update Marks Function

def update_marks():
    # Update student marks
    roll_no = input("Enter Roll Number to update: ").strip()

    for student in students:
        if student["roll_no"] == roll_no:
            print(f"Current Marks: {student['marks']}")
            try:
                new_marks = float(input("Enter new marks: "))
                if new_marks < 0 or new_marks > 100:
                    print("Marks should be between 0-100!")
                    return
                student["marks"] = new_marks
                print("Marks Updated Successfully!")
            except ValueError:
                print("Please enter valid marks!")
            return

    print("Student not found!")

print("update_marks() defined")
```

```python
# Delete Student Function

def delete_student():
    # Delete a student
    roll_no = input("Enter Roll Number to delete: ").strip()

    for i, student in enumerate(students):
        if student["roll_no"] == roll_no:
            name = student["name"]
            students.pop(i)
            print(f"Student '{name}' Deleted Successfully!")
            print(f"Remaining Students: {len(students)}")
            return

    print("Student not found!")

print("delete_student() defined")
```

```python
# Save Data Function

def save_data():
    # Save all student data to file
    if len(students) == 0:
        print("No data to save!")
        return

    try:
        with open("students.txt", "w") as file:
            for student in students:
                line = (f"{student['roll_no']},"
                        f"{student['name']},"
                        f"{student['marks']}\n")
                file.write(line)
        print(f"✓ Data Saved to 'students.txt'!")
        print(f"Total {len(students)} records saved")
    except Exception as e:
        print(f"Error saving data: {e}")

print("save_data() defined")
```

```python
# Load Data Function

def load_data():
    # Load student data from file
    global students
    try:
        with open("students.txt", "r") as file:
            students = []
            for line in file:
                parts = line.strip().split(",")
                if len(parts) == 3:
                    student = {
                        "roll_no": parts[0],
                        "name":    parts[1],
                        "marks":   float(parts[2])
                    }
                    students.append(student)
        print(f"Data Loaded Successfully!")
        print(f"Total {len(students)} students loaded")
    except FileNotFoundError:
        print("No saved file found. Starting fresh!")
    except Exception as e:
        print(f"Error loading data: {e}")

print("load_data() defined")
```

```python
# Main Menu Function

def menu():
    # Main menu
    while True:
        print("\n" + "="*50)
        print("   STUDENT MANAGEMENT SYSTEM")
        print("="*50)
        print("1.Add Student")
        print("2.View Students")
        print("3.Search Student")
        print("4.Update Marks")
        print("5.Delete Student")
        print("6.Save Data")
        print("7.Load Data")
        print("8.Exit")
        print("="*50)

        choice = input("Enter your choice (1-8): ").strip()

        if   choice == "1": add_student()
        elif choice == "2": view_students()
        elif choice == "3": search_student()
        elif choice == "4": update_marks()
        elif choice == "5": delete_student()
        elif choice == "6": save_data()
        elif choice == "7": load_data()
        elif choice == "8":
            print("Goodbye! Thank you!")
            break
        else:
            print("Invalid choice! Try again.")

print("menu() defined")
```

```python

# RUN THE APPLICATION

print("Welcome to Student Management System!")
print("Loading previous data...")
load_data()
menu()
```
